```python
from array import*
tenths = array('i', [10, 20, 30, 40, 50])
for x in tenths:
    print(x)

print("\n")
tenths.insert(3, 35) #insert
for x in tenths:
    print(x)

print("\n")
tenths.remove(50)   #deletion
for x in tenths:
    print(x)

print("\n")
tenths[0] = 19     #update
for x in tenths:
    print(x)
```
